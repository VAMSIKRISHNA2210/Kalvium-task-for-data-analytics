```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv("eci_data_2024_d2_new.csv")
df.head(10)
```

```python
print(df.columns)
print(df.info())
```

```python
df['EVM Votes'] = df['EVM Votes'].astype(str)
df['Postal Votes'] = df['Postal Votes'].astype(str)
df['Total Votes'] = df['Total Votes'].astype(str)
df['% of Votes'] = df['% of Votes'].astype(str)

df['EVM Votes'] = pd.to_numeric(df['EVM Votes'], errors='coerce')
df['Postal Votes'] = pd.to_numeric(df['Postal Votes'], errors='coerce')
df['Total Votes'] = pd.to_numeric(df['Total Votes'], errors='coerce')
df['% of Votes'] = pd.to_numeric(df['% of Votes'].str.rstrip('%'), errors='coerce')

print(df.dtypes)

```

```python
import numpy as np

print(df.isnull().sum())

df.fillna(0, inplace=True)  # Replace NaN with 0
df.replace([np.inf, -np.inf], 0, inplace=True)  # Replace inf with 0

df['EVM Votes'] = df['EVM Votes'].astype(int)
df['Postal Votes'] = df['Postal Votes'].astype(int)
df['Total Votes'] = df['Total Votes'].astype(int)

print(df)
print(df.dtypes)
```

```python
df['Candidate'] = df['Candidate'].str.title()
```

```python
df.head()
```

```python
# Data Exploration
```

```python
# Summary statistics
print("\nSummary statistics:")
print(df.describe())

# Check for any missing values
print("\nMissing values:")
print(df.isnull().sum())

# Check unique parties
print("\nUnique parties:")
print(df['Party'].unique())
```

```python
top_parties = df.groupby('Party')['Total Votes'].sum().sort_values(ascending=False).head(10)
print("Top 10 Parties by Total Votes:")
print(top_parties)
```

```python
plt.figure(figsize=(10,9))
sns.barplot(x=top_parties.index, y=top_parties.values)
plt.xticks(rotation=45)
plt.xlabel('Party')
plt.ylabel('Total Votes')
plt.title('Top 10 Parties by Total Votes')
plt.tight_layout()
plt.show()
```

```python
state_party_votes = df.groupby(['State', 'Party'])['Total Votes'].sum().reset_index()


state_max_party = state_party_votes.loc[state_party_votes.groupby('State')['Total Votes'].idxmax()]


plt.figure(figsize=(12, 8))
sns.barplot(x='State', y='Total Votes', hue='Party', data=state_max_party)
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total Votes')
plt.title('Dominant Parties by State')
plt.legend(title='Party', loc='upper right')
plt.tight_layout()
plt.show()
```

```python

df['Postal Votes'] = pd.to_numeric(df['Postal Votes'], errors='coerce')

state_party_postal_votes = df.groupby(['State', 'Party'])['Postal Votes'].sum().reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='State', y='Postal Votes',data=state_party_postal_votes)
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Postal Votes')
plt.title('Distribution of Postal Votes by State')
#plt.legend(title='Party', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

top_10_highest_margin = df.nlargest(10, 'Margin')

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:d}".format(absolute)

# Plotting a pie chart for the highest margin wins
plt.figure(figsize=(10, 8))
plt.pie(top_10_highest_margin['Margin'], labels=top_10_highest_margin['Constituency'], autopct=lambda pct: func(pct, top_10_highest_margin['Margin']), startangle=140)
plt.title('Top 10 Constituencies with the Highest Margins of Victory')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python

print(df['Margin'].isna().sum())
print(np.isinf(df['Margin']).sum())
df_cleaned = df.replace([np.inf, -np.inf], np.nan).dropna(subset=['Margin'])
df_cleaned = df_cleaned[df_cleaned['Margin'] > 0]
print(df_cleaned['Margin'].describe())
top_10_lowest_margin = df_cleaned.nsmallest(10, 'Margin')


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:d}".format(absolute)


plt.figure(figsize=(10, 8))
plt.pie(top_10_lowest_margin['Margin'], labels=top_10_lowest_margin['Constituency'], autopct=lambda pct: func(pct, top_10_lowest_margin['Margin']), startangle=140)
plt.title('Top 10 Constituencies with the Lowest Margins of Victory')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your Dataset 1 with columns as per your description
# Aggregating total votes for leading and trailing candidates by state
state_candidate_votes = df.groupby(['state', 'candidate'])['total_votes'].sum().reset_index()

# Filtering leading and trailing candidates
leading_candidates = state_candidate_votes.groupby('state').apply(lambda x: x.nlargest(1, 'total_votes')).reset_index(drop=True)
trailing_candidates = state_candidate_votes.groupby('state').apply(lambda x: x.nsmallest(1, 'total_votes')).reset_index(drop=True)

# Merging leading and trailing candidates
merged_candidates = pd.merge(leading_candidates, trailing_candidates, on='state', suffixes=('_leading', '_trailing'))

# Plotting using grouped bar plot
plt.figure(figsize=(14, 8))
bar_width = 0.35

# Positions for bars
r1 = range(len(merged_candidates))
r2 = [x + bar_width for x in r1]

plt.bar(r1, merged_candidates['total_votes_leading'], color='b', width=bar_width, label='Leading Candidate')
plt.bar(r2, merged_candidates['total_votes_trailing'], color='g', width=bar_width, label='Trailing Candidate')

plt.xlabel('State')
plt.ylabel('Total Votes')
plt.title('Competetion based on the difference between Leading and Trailing Candidates by State')
plt.xticks([r + bar_width / 2 for r in range(len(merged_candidates))], merged_candidates['state'], rotation=90)
#plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your Dataset 1 with columns as per your description
# Aggregating total votes for leading and trailing candidates by state
state_candidate_votes = df.groupby(['state', 'candidate'])['total_votes'].sum().reset_index()

# Filtering leading and trailing candidates
leading_candidates = state_candidate_votes.groupby('state').apply(lambda x: x.nlargest(1, 'total_votes')).reset_index(drop=True)
trailing_candidates = state_candidate_votes.groupby('state').apply(lambda x: x.nsmallest(1, 'total_votes')).reset_index(drop=True)

# Merging leading and trailing candidates
merged_candidates = pd.merge(leading_candidates, trailing_candidates, on='state', suffixes=('_leading', '_trailing'))

# Sorting by total votes of the leading candidate in ascending order
merged_candidates.sort_values(by='total_votes_leading', ascending=True, inplace=True)

# Plotting using grouped bar plot
plt.figure(figsize=(14, 8))
bar_width = 0.35

# Positions for bars
r1 = range(len(merged_candidates))
r2 = [x + bar_width for x in r1]

plt.bar(r1, merged_candidates['total_votes_leading'], color='b', width=bar_width, label='Leading Candidate')
plt.bar(r2, merged_candidates['total_votes_trailing'], color='g', width=bar_width, label='Trailing Candidate')
plt.text(0.5, 0.95, "Lowest vote difference shows Highest Competition", ha='center', va='center', transform=plt.gca().transAxes,
         fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
plt.xlabel('State')
plt.ylabel('Total Votes')
plt.title('Competition based on the difference between Leading and Trailing Candidates by State')
plt.xticks([r + bar_width / 2 for r in range(len(merged_candidates))], merged_candidates['state'], rotation=90)
#plt.legend()
plt.tight_layout()
plt.show()
```
